```python
dimensions = df.dimensions.apply(lambda x: tuple(eval(x))).tolist()
print(dimensions)
dimensions[0][2]
n_datapoints = [ (dimensions[i][0]-5) * (dimensions[i][1]-4) * (dimensions[i][2]-4) for i in range(len(dimensions))]
total_datapoints = sum(n_datapoints)
total_datapoints
```

```python
# Creating figure
fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection ="3d")

xs, ys, zs = [],[],[]
for x,y,z in dimensions:
    xs.append(x)
    ys.append(y)
    zs.append(z)
    
# Creating plot
ax.scatter3D(xs, zs, ys, color = "green")
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_zlabel('Y')

plt.title("simple 3D scatter plot")
 
# show plot
plt.show()
```
