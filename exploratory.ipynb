```python
def schem_extractions_to_files(block_arrays, filename_titles, all_mappings, schem_dims):
    #with info, make universal mapping, convert all block arrays, make dataframe of file_of_origin, dest_file, dims; then write the converted block arrays to files
    block_tokens = util.get_blocks_tokenized(all_mappings)
    #make function for converting block array
    tokenized_schematics_blocks = []
    for i, blocks in enumerate(block_arrays):
        converted_blocks = np.array(blocks)
        converted_blocks = util.convert_blocks_to_mapping(blocks, all_mappings[i], block_tokens)
        tokenized_schematics_blocks.append(converted_blocks)

    #create the dataframe with: schematic_filename, tokenized_filename, dimensions
    tokenized_filenames = []
    for title in filename_titles:
        tokenized_filenames.append(title.removesuffix('schematic') + '.tkn')

    df_template = {'schematic_filename': filename_titles, 'tokenized_filenames' : tokenized_filenames, 'dimensions': schem_dims}
    df = pd.DataFrame(df_template)

    #write tokenized block array to files
    tokenized_path = './datasets/preliminary/tokenized_schems/'
    for i , blocks in enumerate(tokenized_schematics_blocks):
        util.write_tokenized_schematic_to_file(blocks, tokenized_path, tokenized_filenames[i])
    #write the dataframe to the parent directory
    
    df.to_csv('./datasets/preliminary/tokenized_schem_data.csv')
    print('Finished writing tokenized files and dataframe')

#--------------------------------------
```

```python
from functools import reduce
def save_ndarray(arr: np.ndarray, path) -> bytes:
    dtype = np.uint8(arr.dtype.num)                                         #dtype stored in 1 byte
    ndims = np.uint8(len(arr.shape))                                        #num dims stored in 1 byte
    dims = np.array([arr.shape[i] for i in range(ndims)], dtype=np.uint32)  #dims stores in 4 bytes each
    
    file_bytes = bytes()
    file_bytes += dtype.tobytes()
    file_bytes += ndims.tobytes()
    file_bytes += dims.tobytes()
    file_bytes += arr.tobytes()
    
    with open(path, 'wb') as file:
        file.write(file_bytes)
    
def load_ndarray(path: str) -> np.ndarray:
    with open(path, 'rb') as file:
        all_bytes = file.read()
        
    dtype = np.sctypeDict[np.frombuffer(all_bytes[0:1], np.uint8)[0]]
    ndims = np.frombuffer(all_bytes[1:2], np.uint8)[0]
    arr_start_idx = 2 + (ndims * np.dtype(np.uint32).itemsize)
    dims = np.frombuffer(all_bytes[2:arr_start_idx], np.uint32)

    size = reduce(lambda x, y: x * y, dims) * np.dtype(dtype).itemsize
    arr = np.frombuffer(all_bytes[arr_start_idx: arr_start_idx + size], dtype=dtype)
    arr.resize(dims)
    return arr
```
